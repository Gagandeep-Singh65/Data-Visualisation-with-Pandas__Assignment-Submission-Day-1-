```python
x = np.arange(1,11)
y = 3 * x + 5
x
```

```python
y
```

```python

plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y, 'go--', linewidth=1, markersize=2)
```

```python
np.pi
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine waveform")

# Plot the points using matplotlib
plt.plot(x, y, color='b')     # color 'b' is blue
plt.show()      # not mandatory line as %matplotlib inline is declared already
```

```python
## pltplot
#plt.plot(x,y, 'ro--')
plt.plot(x,y, 'r*', linestyle='dashed', linewidth=2, markersize=12)   # r* means red color of line and * as marker/point
plt.xlabel('X-axis', color='blue')
plt.ylabel('Y-axis', color='blue')
plt.title('2d-Diagram', color='blue')
```

Creating Subplots

```python
x = np.arange(0,10)
y = 3*x + 10
## plt plot
plt.plot(x,y, 'ro--')
plt.xlabel('X-axis', color='blue')
plt.ylabel('Y-axis', color='blue')
plt.title('2d-Diagram', color='blue')
```

```python

plt.subplot(2,2,1)
plt.plot(x,y, 'r--')
plt.subplot(2,2,2)
plt.plot(x,y, 'g*--')
plt.subplot(2,2,3)
plt.plot(x,y, 'bo--')
plt.subplot(2,2,4)
plt.plot(x,y, 'go--')
```

```python
days = [1,2,3,4,5,6,7]   #days of the week
sales_1 = [160,150,140,145,175,165,180]   #sales of company1
sales_2 = [70,90,160,150,140,145,175]     #sales of company2

plt.figure(figsize = (6,5), dpi=150 )
plt.plot(days, sales_1, 'bo-')
plt.plot(days, sales_2, 'ro-')
plt.xlabel('Days', color='blue')
plt.ylabel('Sales', color='blue')
plt.title('Sales Trend', color='blue')
```
